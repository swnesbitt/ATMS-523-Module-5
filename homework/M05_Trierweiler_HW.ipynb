```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,root_mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('radar_parameters.csv')
```

```python
# calculate Z score form ZH
df['Z'] = 10**(df['Zh (dBZ)']/10)
```

```python
df
```

```python
# creating X and y
X = df[['Zh (dBZ)','Zdr (dB)','Ldr (dB)','Kdp (deg km-1)',
'Ah (dBZ/km)','Adr (dB/km)']]
y = df[['R (mm/hr)']]
```

```python
# calculate rain rate from formula
df['R_formula'] = (df['Z']/200)**(1/1.6)
```

```python
df
```

```python
X
```

```python
y
```

```python
# train test split train size .7
X1, X2, y1, y2 = train_test_split(X, y, random_state=0,
                                  train_size=0.7)
```

```python
model = LinearRegression(fit_intercept=True)
model.fit(X1, y1)
```

```python
X2
```

```python
y_pred_train = model.predict(X1)
y_pred_test = model.predict(X2)
```

```python
lr_r2_train = r2_score(y1, y_pred_train)
lr_r2_test = r2_score(y2, y_pred_test)
lr_rmse_train = root_mean_squared_error(y1, y_pred_train)
lr_rmse_test = root_mean_squared_error(y2, y_pred_test)
print("Linear Regression R^2 training set: {}".format(lr_r2_train))
print("Linear Regression R^2 testing set: {}".format(lr_r2_test))
print("Linear Regression RMSE training set: {}".format(lr_rmse_train))
print("Linear Regression RMSE testing set: {}".format(lr_rmse_test))
```

```python
# R^2 Rain Rate against formula rain rate
f_r2 = r2_score(df['R (mm/hr)'],df['R_formula'])
print("Formula Vs Actual rain rate R^2: {}".format(f_r2))
```
